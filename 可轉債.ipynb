$CB_t=e^{-r(T-t)}F+qS_tN(d_1)-Fe^{-r(T-t)}N(d_2)$

$d_1=\frac{\ln(\frac{qS_t}{F})+(r+\frac{1}{2}\sigma^2)(T-t)}{\sigma\sqrt{T-t}}$

$d_2=\frac{\ln(\frac{qS_t}{F})+(r-\frac{1}{2}\sigma^2)(T-t)}{\sigma\sqrt{T-t}}=d_1-\sigma\sqrt{T-t}$

其中$CB_t$為可轉債在$t$時間點的價格；$r$為無風險利率；$T$為到期日時機點；$t$為目前時間點；$F$為債券面額；$q$為轉換比率；$\sigma$為股票報酬波動度；$S_t$為股票在$t$的價格。

```python
def bond_price(S, F, q, r, sigma, T):
  d1 = (math.log(q*S/F)+(r+0.5*sigma**2)*T)/(sigma*math.sqrt(T))
  d2 = d1-sigma*math.sqrt(T)
  price = math.exp(-r*T)*F + q*S*norm.cdf(d1) - F*math.exp(-r*T)*norm.cdf(d2)
  return price

import math
from scipy.stats import norm
from datetime import datetime

maturity = datetime.strptime("2023-12-17", "%Y-%m-%d") #到期日
data['T'] = (maturity - data['Date']).dt.days/ 365 #距到期日

F = 100 #面額，通常為100
q = 0.0001 #轉換比率
r = 0.0108 #無風險利率
sigma = 0.3068 #股價波動度

thm = list(map(lambda S, T: bond_price(S, F, q, r, sigma, T), data['Stock price'], data['T']))
data['Theoretical price'] = thm
data['Error rate'] = round((abs(data['Theoretical price'] - data['Market price'])/ data['Market price'])*100, 2)
data.to_csv("/content/drive/MyDrive/政大課程/111財務數學/可轉債/result.csv")
```

```python
import matplotlib as mpl

fig, ax1 = plt.subplots(figsize=(10,5))
# label_size = 8
mpl.rcParams['xtick.labelsize'] = 14 
mpl.rcParams['ytick.labelsize'] = 14 

ax1.set_xlabel('Date', fontsize='16')
ax1.set_ylabel('Price', fontsize='16')
f1 = ax1.plot(data['Date'], data['Market price'], marker='^', label='Market price',ms=10)
f2 = ax1.plot(data['Date'], data['Theoretical price'], marker='o', label='Theoretical price',ms=10)

plt.xticks(rotation=45)
plt.grid(axis='x')

ax2 = ax1.twinx()
color = 'tab:green'
ax2.set_ylabel('Error rate(%)', color=color, fontsize='16')

f3 = ax2.plot(data['Date'], data['Error rate'], color=color, marker='s',ms=10,label='Error rate(%)')
ax2.set_yticklabels(data['Error rate'], fontsize=14)
fig.tight_layout()


# added these three lines
lns = f1+f2+f3
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, fontsize=12, loc='center left', bbox_to_anchor=(1.11, 0.5),shadow=True, ncol=1)


plt.savefig("/content/drive/MyDrive/政大課程/111財務數學/可轉債/error.png", bbox_inches='tight')
               
plt.show()
```
